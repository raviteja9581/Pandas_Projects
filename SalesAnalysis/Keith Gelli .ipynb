```python
import pandas as pd
import os 
import glob 
```

```python
df= pd.DataFrame()
filepath= "./Sales_Data/"
files= glob.glob(filepath+'/*.csv')
for file in files:
    tempfile= pd.read_csv(file)
    df = df.append(tempfile,ignore_index=True)
```

```python
df.info()
```

```python
f= df['Order ID'].str.contains("NAN",na=True)
```

```python
df.info()
```

```python
# Removing the NAN 
df=df.dropna()
```

```python
df[f]
```

```python
for x in df.index:
    if df.loc[x,'Order ID'] == 'Order ID':
        df.drop(x,inplace=True)
```

```python
df.info()
```

```python
df['Month']=df['Order Date'].str[0:2]
```

```python
df.info()
```

```python
df['Quantity Ordered']=df['Quantity Ordered'].astype('float32')
```

```python
df.info()
```

```python
df['Price Each']=df['Price Each'].astype('float32')
```

```python
df.info()
```

```python
df['Sale_value']= df['Quantity Ordered']*df['Price Each']
```

```python
df.info()
```

```python
df
```

```python
f2= df['Month'].str.contains('04',na=True)
```

```python
df[f2]
```

```python
result=df.groupby('Month').sum()['Sale_value']
```

```python
import matplotlib.pyplot as plt
```

```python
months=range(1,13)
plt.bar(months,result)
plt.xticks(months)
plt.ylabel('Sales in US Dollar ($)')
plt.xlabel('Month Number')
plt.show()
```

# What time we should advertise to maximize the sales

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Hour']= df['Order Date'].dt.hour
```

```python
df['Minute'] = df['Order Date'].dt.minute
```

```python
df
```

```python
hours = [hour for hour, df2 in df.groupby('Hour')]
```

```python
plt.plot(hours, df.groupby('Hour').count())
plt.xticks(hours)
plt.grid()
plt.ylabel("No of Orders")
plt.xlabel('Hours')
plt.show()
```
